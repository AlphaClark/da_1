#Importing Required Libraries

```python
import numpy as np
import pandas as pd
```

#Converting Dataset to Pandas Dataframe to perform operations

```python
df = pd.read_csv("/content/netflix-shows/netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

#Handling Null Values

```python
df.dropna(subset=['director','cast','country','rating'], inplace=True)
for col in ['date_added', 'duration']:
  df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum()
```

#Removing Duplicate Values

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

#Standardising Country Names

```python
country_mapping = {
    'United States': 'USA',
    'United Kingdom': 'UK',
}
df['country'] = df['country'].replace(country_mapping)
print(df['country'].value_counts())
```

#Converting Dates to a Consistent Format

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y').dt.strftime('%d-%m-%Y')
df.head()
```

#Renaming Column Headers for Uniformity

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```
